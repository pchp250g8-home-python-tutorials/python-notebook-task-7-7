Найти сумму всех трёхзначных чисел, делящихся на 7.

```python
from sympy import Eq
from sympy import solve
from sympy import symbols
from sympy import log

a1 = 105; an = 994; d = 7
n = symbols("n")
eq1 = Eq(a1 + d*(n - 1), an)
print(eq1.lhs,"=",eq1.rhs)
r1 = solve(eq1,n);
n1 = r1[0]
s = (2 * a1 + d * (n1 - 1)) * n1 / 2
print([n1, s])
```
